# MATCH

```python
def http_error(status):
    match status:
        case 400:
            return "BAD REQUEST"
        case 401 | 403 :
            return "NOT ALLOWED"
        case 404:
            return "NOT FOUND"
        case _:
            raise ValueError("UNDEFINED ERROR")
```

```python
http_error(400)
```

```python
http_error(403)
```

```python
http_error(404)
```

```python
http_error(200)
```

FUNCTION

```python
def fibo(n):
    a, b = 0, 1
    while a < n:
        print(a, end=' ')
        a, b = b, a + b
    print()
```

```python
fibo(100)
```

```python
def fibo2(n):
    result = []
    a, b = 0, 1
    while a < n:
        result.append(a)
        a, b = b, a + b
    return result
```

```python
fibo_result = fibo2(100)
```
