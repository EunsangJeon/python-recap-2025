# Q3

```python
bike_q3 = pd.read_csv('bike.csv')
```

```python
bike_q3
```

```python
bike_q3_sub = bike_q3
bike_q3_sub['datetime'] = pd.to_datetime(bike_q3_sub['datetime'])
bike_q3_sub
```

```python
bike_q3_sub['hour'] = bike_q3_sub['datetime'].dt.hour
```

```python
bike_q3_sub['day'] = bike_q3_sub['datetime'].dt.date
```

```python
bike_q3_sub
```

```python
bike_q3_sub.groupby('day')['humidity'].max()
```

```python
rainy_day = bike_q3_sub.groupby('day')['humidity'].max().reset_index()
```

```python
rainy_day
```

```python
rainy_day = rainy_day[rainy_day['humidity'] == 100]
```

```python
rainy_day
```

```python
len(rainy_day)
```

```python
q3_filtered = pd.merge(left=bike_q3_sub, right=rainy_day, left_on='day', right_on='day', how='inner')
```

```python
q3_filtered.info()
```

```python
over_30 = q3_filtered[q3_filtered['temp'] > 30]
```

```python
over_30
```

```python
over_30['count'].mean()
```
