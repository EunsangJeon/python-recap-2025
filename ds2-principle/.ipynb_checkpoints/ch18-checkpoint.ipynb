```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv('iris.csv')
```

```python
df.head()
```

```python
df['Species'].unique()
```

```python
df['Species'].value_counts()
```

```python
df['is_versicolor'] = (df['Species'] == 'versicolor') + 0
```

```python
df.head()
```

```python
model_c = DecisionTreeClassifier(random_state=123, max_depth=5).fit(y=df['is_versicolor'], X=df.iloc[:, :4])
```

```python
pred_c = model_c.predict(df.iloc[:, :4])
```

```python
pred_c
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_true=df['is_versicolor'], y_pred=pred_c)
```

```python
model_r = DecisionTreeRegressor(random_state=123).fit(y=df['Petal.Width'], X=df.iloc[:, :3])
```

```python
pred_r = model_r.predict(df.iloc[:, :3])
```

```python
pred_r
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_true=df['Petal.Width'], y_pred=pred_r)
```

# Q1

```python
import pandas as pd
```

```python
df_q1 = pd.read_csv('diabetes.csv')
df_q1.head()
```

```python
df_q1 = df_q1.loc[:, ['Glucose', 'BloodPressure', 'Pregnancies', 'Outcome']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
q1_train, q1_test = train_test_split(df_q1, random_state=123, train_size=0.8)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model_q1 = DecisionTreeClassifier(random_state=123).fit(X=q1_train.iloc[:, :3], y=q1_train['Outcome'])
```

```python
pred_q1 = model_q1.predict(q1_test.iloc[:, :3])
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_true=q1_test['Outcome'], y_pred=pred_q1)
```

```python
model_q2 = DecisionTreeRegressor(random_state=123).fit(y=q2_train['BMI'], X=q2_train.iloc[:, :3])
```

```python
from sklearn.metrics import mean_squared_error
```

```python
round(mean_squared_error(y_true=q2_test['BMI'], y_pred=pred_q2) ** 0.5, 1)
```

# Q3

```python
import pandas as pd
```

```python
df_q3 = pd.read_csv('diabetes.csv')
df_q3.head()
```

```python
df_q3.columns
```

```python
df_q3 = df_q3.loc[:, ['Glucose', 'BloodPressure', 'Pregnancies', 'BMI', 'Age','Outcome']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
q3_train, q3_test = train_test_split(df_q3, train_size=0.7, random_state=345)
```

```python
dep = [3, 4, 5, 6]
```

```python
res = pd.DataFrame(dep, columns=['d'])
```

```python
res
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
acc = []
for d in dep:
    model = DecisionTreeClassifier(max_depth=d, random_state=345).fit(y=q3_train['Outcome'], X=q3_train.iloc[:, :5])
    pred = model.predict(q3_test.iloc[:, :5])
    acc.append(round(accuracy_score(y_true=q3_test['Outcome'], y_pred=pred), 2))
res['acc'] = acc
```
