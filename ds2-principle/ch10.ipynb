```python
import pandas as pd
```

```python
def udf_sum(x, y):
    return x + y;
```

```python
udf_sum(pd.Series([2, 4, 6]), 4)
```

```python
def udf_sum2(x, y):
    return x.sum() + y
```

```python
udf_sum2(pd.Series([2, 4, 6]), 4)
```

```python
def udf_mean1(x):
    return round(x.mean(), 3)
```

```python
udf_mean1(pd.Series([4, 6, 8, 17 ,23, 19, 13]))
```

```python
def udf_mean2(x, digit):
    return round(x.mean(), digit)
```

```python
udf_mean2(pd.Series([4, 6, 8, 17 ,23, 19, 13]), 4)
```

```python
def udf_mean3(x, digit=3):
    return round(x.mean(), digit)
```

```python
udf_mean3(pd.Series([4, 6, 8, 17 ,23, 19, 13]))
```

# Q2

```python
def q2(a):
    return (a - a.mean()) / a.std()
```

```python
q2(pd.Series([-4, 5, 7, 9]))
```

# Q3

```python
def q3(a):
    return (a - a.min()) / (a.max() - a.min())
```

```python
q3(pd.Series([-4, 5, 7, 9]))
```
