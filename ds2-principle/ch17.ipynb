```python
import pandas as pd
```

```python
df_q1 = pd.read_csv('diabetes.csv')
df_q1.head()
```

```python
df_q1 = df_q1.loc[:, ['Pregnancies', 'Glucose', 'BloodPressure', 'Outcome']]
```

```python
df_q1_train, df_q1_test = train_test_split(df_q1, train_size=0.7, random_state=123)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model_q1 = KNeighborsClassifier().fit(y=df_q1_train['Outcome'], X=df_q1_train.iloc[:, :3])
```

```python
pred_q1 = model_q1.predict(df_q1_test.iloc[:, :3])
```

```python
pred_q1
```

```python
from sklearn.metrics import accuracy_score
```

```python
round(accuracy_score(y_true=df_q1_test['Outcome'], y_pred=pred_q1), 2)
```

# Q3

```python
import pandas as pd
```

```python
num_k = [3, 5, 10, 20]
result = pd.DataFrame()
```

```python
result['k'] = num_k
```

```python
df_q3 = pd.read_csv('diabetes.csv')
df_q3.head()
```

```python
df_q3['is_preg'] = (df_q3['Pregnancies'] > 0) + 0
```

```python
df_q3 = df_q3.loc[:, ['is_preg', 'Glucose', 'BloodPressure', 'Insulin', 'BMI']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
q3_train, q3_test = train_test_split(df_q3, train_size=0.8, random_state=123)
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
```

```python
rmses = []
for i in range(len(num_k)):
    model = KNeighborsRegressor(n_neighbors=num_k[i]).fit(y=q3_train['BMI'], X=q3_train.iloc[:, :4])
    pred = model.predict(X=q3_test.iloc[:, :4])
    rmse = mean_squared_error(y_true=q3_test['BMI'], y_pred=pred) ** 0.5
    rmses.append(rmse)
    
result['rsme'] = rmses
```
