```python
import pandas as pd
from sklearn.naive_bayes import GaussianNB
```

```python
df = pd.read_csv('iris.csv')
df.head()
```

```python
df['is_setosa'] = (df['Species'] == 'setosa') + 0
df.head()
```

```python
df.tail()
```

```python
df['is_setosa'].unique()
```

```python
df['is_setosa'].value_counts()
```

```python
df['is_setosa'].value_counts(normalize=True)
```

```python
model = GaussianNB().fit(X=df.iloc[:,:4], y=df['is_setosa'])
model
```

```python
model.class_prior_
```

```python
model.theta_
```

```python
model.predict_proba(df.iloc[:, :4])
```

```python
pred = model.predict_proba(df.iloc[:, :4])
pred = pred[:,1]
pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
pred = (pred > 0.9) + 0
```

```python
pred[:4]
```

```python
accuracy_score(y_true=df['is_setosa'], y_pred=pred)
```

# Q3

```python
import pandas as pd
```

```python
df_q3 = pd.read_csv('diabetes.csv')
df_q3.head()
```

```python
df_q3['is_preg'] = (df_q3['Pregnancies'] > 0) + 0
```

```python
df_q3['age_ran'] = (df_q3['Age'] // 10) * 10
```

```python
df_q3.head()
```

```python
df_q3 = df_q3.loc[:, ['is_preg', 'age_ran', 'BMI', 'Glucose', 'Outcome']]
df_q3 = df_q3.loc[df_q3['BMI'] > 0]
```

```python
df_q3.head()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
df_q3_train, df_q3_test = train_test_split(df_q3, train_size=0.8, random_state=123)
```

```python
q3_logi = LogisticRegression(random_state=123).fit(X=df_q3_train.iloc[:, :4], y=df_q3_train.iloc[:, 4])
```

```python
pred_logi = q3_logi.predict_proba(df_q3_test.iloc[:, :4])
```

```python
pred_logi = pred_logi[:, 1]
```

```python
pred_logi = (pred_logi > 0.5) + 0
```

```python
accr_logi = accuracy_score(y_true=df_q3_test.iloc[:, 4], y_pred=pred_logi)
```

```python
accr_logi
```

```python
q3_nb = GaussianNB().fit(X=df_q3_train.iloc[:, :4], y=df_q3_train.iloc[:, 4])
```

```python
pred_nb = q3_nb.predict(df_q3_test.iloc[:, :4])
```

```python
# pred_nb = pred_nb[:, 1]
```

```python
# pred_nb = (pred_nb > 0.5) + 0
```

```python
accr_nb = accuracy_score(y_true=df_q3_test.iloc[:, 4], y_pred=pred_nb)
```
