```python
df_q3 = pd.read_csv('iris.csv')
```

```python
df_q3.head()
```

```python
q3_sl_mean = df_q3['Sepal.Length'].mean()
q3_sl_std = df_q3['Sepal.Length'].std()
print(q3_sl_mean)
print(q3_sl_std)
```

```python
cond1 = df_q3['Sepal.Length'] < (q3_sl_mean - 1.5 * q3_sl_std)
cond2 = df_q3['Sepal.Length'] > (q3_sl_mean + 1.5 * q3_sl_std)
```

```python
cond1
```

```python
df_out = df_q3.loc[cond1 | cond2]
```
