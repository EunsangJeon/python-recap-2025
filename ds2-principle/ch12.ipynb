```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
```

```python
df_q2 = pd.read_csv('diabetes.csv')
df_q2.head()
```

```python
df_q2 = df_q2[df_q2['BMI'] != 0]
```

```python
df_q2_nor = MinMaxScaler().fit(df_q2).transform(df_q2)
```

```python
df_q2_nor
```

```python
df_q2_nor = pd.DataFrame(df_q2_nor, columns=df_q2.columns)
```

```python
df_q2_nor.head()
```

```python
df_q2_kmeans = KMeans(n_clusters=4, random_state=123).fit(df_q2_nor)
```

```python
df_q2_kmeans.labels_
```

```python
df_q2['label'] = df_q2_kmeans.labels_
```

```python
df_q2['label'].value_counts()
```

```python
df_q2[df_q2['label'] == 1]
```

```python
df_q2[df_q2['label'] == 1]['Age'].mean().round(0)
```

```python
import pandas as panda
from sklearn.cluster import KMeans
```

```python
df_q3 = pd.read_csv('diabetes.csv')
df_q3.head()
```

```python
df_q3 = df_q3[df_q3['BMI'] != 0].reset_index(drop=True)
```

```python
df_q3_centres = pd.DataFrame(KMeans(n_clusters=3, random_state=123).fit(df_q3).cluster_centers_, columns=df_q3.columns)
```

```python
df_q3['label'] = KMeans(n_clusters=3, random_state=123).fit(df_q3).labels_
```

```python
df_q3.head()
```

```python
df_q3_centres
```

```python
for i in range(3):
    for j in range(i, 3):
        if i == j:
            continue
        print(i, j, ((df_q3_centres.iloc[i,] - df_q3_centres.iloc[j,]) ** 2).sum() ** 0.5)
```
