# Q2

```python
import pandas as pd
from scipy.stats import pearsonr
```

```python
df_q2 = pd.read_csv('bike.csv')
df_q2.head()
```

```python
df_q2_sum = df_q2[['season', 'atemp', 'casual']]
```

```python
for i in range(1, 5):   
    print(df_q2_sum[df_q2_sum['season'] == i][['atemp', 'casual']].corr().round(3))
```

```python
df_corr = df_q2[['season', 'atemp', 'casual']].groupby('season').corr().reset_index()
```

```python
df_corr.head()
```

```python
df_corr = df_corr[df_corr['level_1'] == 'casual'].reset_index()
```

```python
df_corr.head()
```

```python
df_corr.sort_values('atemp')
```
