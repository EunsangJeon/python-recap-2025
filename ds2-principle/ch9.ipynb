```python
import pandas as pd
```

```python
dia = pd.read_csv('diamonds.csv')
```

```python
dia.head()
```

```python
pd.crosstab(dia['cut'], dia['clarity'])
```

```python
dia_sub = dia.loc[(dia['cut'] == 'Good') & (dia['clarity'] == 'IF')]
len(dia_sub)
```

```python
sum((dia['cut'] == 'Good') & (dia['clarity'] == 'IF'))
```

```python
pd.crosstab(dia['cut'], dia['clarity'], normalize=True)
```

```python
pd.crosstab(dia['cut'], dia['clarity'], normalize=True).round(2)
```

```python
(pd.crosstab(dia['cut'], dia['clarity'], normalize=True) * 100).round(2)
```

```python
pd.crosstab(dia['cut'], dia['clarity'], values=dia['price'], aggfunc='mean')
```

```python
dia.groupby(['cut', 'clarity'])['price'].mean().round(2).reset_index()
```

```python
dia_agg = dia.groupby(['cut', 'clarity'])['price'].mean().reset_index()
```

```python
dia_agg.head()
```

```python
dia_agg.sort_values('price')
```

```python
dia_agg.sort_values('price', ascending=False)
```

```python
dia_agg.sort_values(['cut', 'price'], ascending=[True, False])
```

```python
dia_q3 = pd.read_csv('diamonds.csv')
```

```python
dia_q3.head()
```

```python
dia_q3['price_per_carat'] = dia_q3['price'] / dia_q3['carat']
```

```python
dia_q3.head()
```

```python
summary_q3 = dia_q3.groupby(['cut', 'color'])['price_per_carat'].mean().reset_index()
```

```python
summary_q3.head()
```

```python
summary_q3.iloc[summary_q3['price_per_carat'].idxmax()]
```

```python
summary_q3.sort_values('price_per_carat', ascending=False)
```
