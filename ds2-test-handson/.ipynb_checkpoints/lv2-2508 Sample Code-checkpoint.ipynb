```python
import pandas as pd
import numpy as np
```

```python
JOB = pd.read_csv('lv2-2508.csv')
JOB.info()
```

```python
print(np.sort(JOB['JOB_CATEGORY'].unique()))
print(np.sort(JOB['EXP_LEVEL'].unique()))
print(np.sort(JOB['WORK_SETTING'].unique()))
print(np.sort(JOB['COMP_SIZE'].unique()))
```

```python
JOB.groupby(['EXP_LEVEL']).agg(CNT = ('E_ID','size'), SALARY = ('SALARY','mean')).reset_index()
```

# 21

21-1

```python
DF11 = JOB[JOB['JOB_CATEGORY']=='A'].copy()
DF11.shape
```

21-2

```python
sal_q1 = DF11['SALARY'].quantile(0.25)
sal_q3 = DF11['SALARY'].quantile(0.75)
```

21-3

```python
DF12 = DF11.groupby(['EXP_LEVEL','WORK_SETTING','COMP_SIZE'])\
.agg(E_CNT=('E_ID', 'size'), SAL_AVG=('SALARY','mean')).reset_index()
DF12.shape
```

21-4

```python
DF_HIGH = DF12[DF12['SAL_AVG']>=sal_q3].copy()
DF_LOW = DF12[DF12['SAL_AVG']<=sal_q1].copy()
```

```python
def avg_(DF_HIGH):
    g_avg = (DF_HIGH['SAL_AVG']*DF_HIGH['E_CNT']).sum()/DF_HIGH['E_CNT'].sum()
    return g_avg
```

```python
res21 = np.abs(avg_(DF_HIGH) - avg_(DF_LOW))
ans21 = round(res21, 1)
print(res21)
print('ans21:', ans21)
```

# 22

```python
df220 = JOB.copy()
```

22-1

```python
df220['EXP'] = 'H'
df220.loc[df220['EXP_LEVEL'].isin(['Entrylevel','Midlevel']), 'EXP'] = 'L'
```

22-2

```python
list_J = list(np.sort(df220['JOB_CATEGORY'].unique()))
list_coef = []

for j in list_J:
    n_h = np.sum((df220['JOB_CATEGORY']==j) & (df220['EXP']=='H'))
    n_l = np.sum((df220['JOB_CATEGORY']==j) & (df220['EXP']=='L'))
    m_h = df220.loc[(df220['JOB_CATEGORY']==j) & (df220['EXP']=='H'), 'SALARY'].mean()
    m_l = df220.loc[(df220['JOB_CATEGORY']==j) & (df220['EXP']=='L'), 'SALARY'].mean()
    s_h = df220.loc[(df220['JOB_CATEGORY']==j) & (df220['EXP']=='H'), 'SALARY'].std()
    s_l = df220.loc[(df220['JOB_CATEGORY']==j) & (df220['EXP']=='L'), 'SALARY'].std()
    
    list_coef.append(np.abs(m_h-m_l)/(((n_h*s_h)+(n_l*s_l))/(n_h+n_l)))
```

22-3

```python
res22 = np.abs(np.max(list_coef) - np.min(list_coef))
ans22 = round(res22, 3)
print(res22)
print('ans22:', ans22)
```

# 23

```python
DF30 = JOB.copy()
```

23-1

```python
df_dum = pd.get_dummies(DF30.loc[:, ['JOB_CATEGORY','EXP_LEVEL','WORK_SETTING','COMP_SIZE']],\
              drop_first=True)
list_dum = list(df_dum.columns)
DF31 = pd.concat([DF30, df_dum], axis=1)
DF31
```

23-2

```python
train30 = DF31[DF31['E_ID']%5!=0].copy()
test30 = DF31[DF31['E_ID']%5==0].copy()
print(train30.shape, test30.shape)
```

23-3

```python
train31 = train30.copy()
```

```python
q1 = train31['SALARY'].quantile(0.25)
q3 = train31['SALARY'].quantile(0.75)
iqr = q3 - q1
train31 = train31[(train31['SALARY']>=q1 - 1.5*iqr) & (train31['SALARY']<=q3 + 1.5*iqr)]
train31.shape
```

23-4

```python
from sklearn.tree import DecisionTreeRegressor
dt00 = DecisionTreeRegressor(random_state=1234, max_depth=4, min_samples_split=100)
dt00.fit(train31.loc[:, list_dum], train31['SALARY'])
```

23-5

```python
arr_pred = dt00.predict(test30.loc[:, list_dum])
arr_act = np.array(test30['SALARY'])
res23 = np.sqrt(np.mean((arr_pred-arr_act)**2))
ans23 = round(res23, 2)
print(res23)
print('ans23:', ans23)
```
