# Q1

```python
import pandas as pd
```

```python
df = pd.read_csv('lv2-2305-1.csv')
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
# step 1-1

scale_target = ['NUM_LOGIN', 'NUM_CONTENTS', 'NUM_ACTION', 'TIME_LOGIN']

for target in scale_target:
    df[target + '_S'] = StandardScaler().fit_transform(df.loc[:, [target]])
```

```python
df.head()
```

```python
# step 1-2

df['INVOLVEMENT'] = df['NUM_LOGIN_S'] + df['NUM_CONTENTS_S'] + df['NUM_ACTION_S'] + df['TIME_LOGIN_S']
```

```python
df.head()
```

```python
# step 1-3

from scipy.stats import pearsonr
```

```python
coef, p = pearsonr(x=df['INVOLVEMENT'], y=df['SCORE'])
```

```python
print(coef)
```

```python
print(round(coef, 3))
```

# Q2

```python
df_2 = df.copy()
# df_2 = df_2.loc[:, ['NUM_LOGIN', 'NUM_CONTENTS', 'NUM_ACTION', 'TIME_LOGIN', 'TIME_TEST', 'NUM_REVISION', 'SCORE']]
```

```python
df_2.head()
```

```python
# 2-1

q1 = df_2['SCORE'].quantile(q=0.25)
```

```python
q1
```

```python
q3 = df_2['SCORE'].quantile(q=0.75)
```

```python
q3
```

```python
iqr= q3 - q1
```

```python
iqr
```

```python
df_2 = df_2[df_2['SCORE'] > q1 - 1.5 * iqr]
```

```python
df_2 = df_2[df_2['SCORE'] < q3 + 1.5 * iqr]
```

```python
# 2-2
df_2_train = df_2[(df_2['EMP_ID'].str[4].astype(int)) % 4 != 0]
```

```python
df_2_test = df_2[(df_2['EMP_ID'].str[4].astype(int)) % 4 == 0]
```

```python
# 2-3

from sklearn.linear_model import LinearRegression
```

```python
df_2_train = df_2_train.loc[:, ['NUM_LOGIN', 'NUM_CONTENTS', 'NUM_ACTION', 'TIME_LOGIN', 'TIME_TEST', 'NUM_REVISION', 'SCORE']]
df_2_train.head()
```

```python
df_2_test = df_2_test.loc[:, ['NUM_LOGIN', 'NUM_CONTENTS', 'NUM_ACTION', 'TIME_LOGIN', 'TIME_TEST', 'NUM_REVISION', 'SCORE']]
```

```python
test_target = ['NUM_LOGIN', 'NUM_CONTENTS', 'NUM_ACTION', 'TIME_LOGIN', 'TIME_TEST', 'NUM_REVISION']
```

```python
df_2_train.shape[0]
```

```python
q2_res =pd.DataFrame(data=['NUM_LOGIN', 'NUM_CONTENTS', 'NUM_ACTION', 'TIME_LOGIN', 'TIME_TEST', 'NUM_REVISION'], columns=['col'])
```

```python
addr_sqrs = []
for target in test_target:
    model = LinearRegression().fit(X = df_2_train[[target]], y = df_2_train['SCORE'])
    r_sqr = model.score(X = df_2_train[[target]], y = df_2_train['SCORE'])
    addr_sqr= 1 - (((1 - r_sqr) * (df_2_train.shape[0] - 1))/(df_2_train.shape[0] - 1 - 1))
    addr_sqrs.append(addr_sqr)
```

```python
addr_sqrs
```

```python
q2_res['addr'] = addr_sqrs
```

```python
q2_res
```

```python
q2_res['addr'].max() # TIME_LOGIN
```

```python
q2_res =pd.DataFrame(data=['NUM_LOGIN', 'NUM_CONTENTS', 'NUM_ACTION', 'TIME_TEST', 'NUM_REVISION'], columns=['col'])
```

```python
test_target = ['NUM_LOGIN', 'NUM_CONTENTS', 'NUM_ACTION', 'TIME_TEST', 'NUM_REVISION']
```

```python
addr_sqrs = []
for target in test_target:
    model = LinearRegression().fit(X = df_2_train.loc[:, ['TIME_LOGIN', target]], y = df_2_train['SCORE'])
    r_sqr = model.score(X = df_2_train.loc[:, ['TIME_LOGIN', target]], y = df_2_train['SCORE'])
    addr_sqr= 1 - (((1 - r_sqr) * (df_2_train.shape[0] - 1))/(df_2_train.shape[0] - 2 - 1))
    addr_sqrs.append(addr_sqr)
addr_sqrs
```

```python
q2_res['addr'] = addr_sqrs
q2_res
```

```python
q2_res['addr'].max() # NUM_CONTENTS
```

```python
q2_res =pd.DataFrame(data=['NUM_LOGIN', 'NUM_ACTION', 'TIME_TEST', 'NUM_REVISION'], columns=['col'])
test_target = ['NUM_LOGIN', 'NUM_ACTION', 'TIME_TEST', 'NUM_REVISION']

addr_sqrs = []
for target in test_target:
    model = LinearRegression().fit(X = df_2_train.loc[:, ['TIME_LOGIN', 'NUM_CONTENTS', target]], y = df_2_train['SCORE'])
    r_sqr = model.score(X = df_2_train.loc[:, ['TIME_LOGIN', 'NUM_CONTENTS', target]], y = df_2_train['SCORE'])
    addr_sqr= 1 - (((1 - r_sqr) * (df_2_train.shape[0] - 1))/(df_2_train.shape[0] - 3 - 1))
    addr_sqrs.append(addr_sqr)
addr_sqrs
```

```python
q2_res['addr'] = addr_sqrs
```

```python
q2_res
```

```python
q2_res['addr'].max() # NUM_REVISION
```

```python
model = LinearRegression().fit(X = df_2_train.loc[:, ['TIME_LOGIN', 'NUM_CONTENTS', 'NUM_REVISION']], y = df_2_train['SCORE'])
```

```python
pred = model.predict(X=df_2_test.loc[:, ['TIME_LOGIN', 'NUM_CONTENTS', 'NUM_REVISION']])
```

```python
from sklearn.metrics import mean_squared_error
```

```python
round(mean_squared_error(y_true = df_2_test['SCORE'], y_pred = pred) ** 0.5, 2)
```

### 3-1

```python
import pandas as pd
```

```python
df_3 = pd.read_csv('lv2-2305-1.csv')
df_3.head()
```

```python
df_a = df_3.loc[df_3['SCORE'] <= 5]
```

```python
df_b = df_3.loc[df_3['SCORE'] > 5]
df_b = df_b.loc[df_b['SCORE'] <= 7]
```

```python
df_c = df_3.loc[df_3['SCORE'] > 7]
df_c = df_c.loc[df_c['SCORE'] <= 10]
```

### 3-2

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
test_col = ['NUM_LOGIN', 'NUM_CONTENTS', 'NUM_ACTION', 'TIME_LOGIN', 'TIME_TEST', 'NUM_REVISION']
```

```python
model_a = RandomForestRegressor(n_estimators=10, random_state=1234, min_samples_leaf=10).fit(X=df_a.loc[:, test_col], y=df_a['SCORE'])
```

```python
model_b = RandomForestRegressor(n_estimators=10, random_state=1234, min_samples_leaf=10).fit(X=df_b.loc[:, test_col], y=df_b['SCORE'])
```

```python
model_c = RandomForestRegressor(n_estimators=10, random_state=1234, min_samples_leaf=10).fit(X=df_c.loc[:, test_col], y=df_c['SCORE'])
```

```python
model_a.feature_importances_, model_b.feature_importances_, model_c.feature_importances_
```

```python
a_fi = pd.DataFrame(test_col, columns=['feat'])
```

```python
b_fi = pd.DataFrame(test_col, columns=['feat'])
```

```python
c_fi = pd.DataFrame(test_col, columns=['feat'])
```

### 3-3

```python
a_fi['fi'] = model_a.feature_importances_
```

```python
b_fi['fi'] = model_b.feature_importances_
```

```python
c_fi['fi'] = model_c.feature_importances_
```

```python
a_fi.iloc[a_fi['fi'].idxmax()]
```

```python
b_fi.iloc[b_fi['fi'].idxmax()]
```
