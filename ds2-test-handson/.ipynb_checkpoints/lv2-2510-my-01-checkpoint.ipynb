# Q 23

# Step 23-1

```python
COMPONENT = pd.read_csv('lv2-2510.csv')
```

```python
COMPONENT.shape
```

```python
COMPONENT.head()
```

```python
COMPONENT.columns
```

```python
VAR3 = ['AREA', 'PERIM', 'MAJOR_LEN', 'MINOR_LEN', 'ASPECT', 'SOLID']
```

```python
len(VAR3)
```

```python
COMPONENT['OUT'] = COMPONENT['COMP_ID'].map(lambda x: 0)
```

```python
COMPONENT.head()
```

```python
for target in VAR3:
    q1 = COMPONENT[target].quantile(q=0.25)
    q3 = COMPONENT[target].quantile(q=0.75)
    iqr = q3-q1
    for i in range(len(COMPONENT[target])):
        if (COMPONENT.loc[i, target] < q1 - iqr * 1.5) | (COMPONENT.loc[i, target] > q3 + iqr * 1.5):
            COMPONENT.loc[i, 'OUT'] += 1
```

```python
COMPONENT.shape
```

```python
COMPONENT['OUT'].value_counts()
```

```python
DF3 = COMPONENT[COMPONENT['OUT'] < 2]
```

```python
DF3.shape
```

```python
DF3 = DF3.reset_index()
```

# Step 23-2

```python
TRAIN3 = DF3[DF3['COMP_ID'] % 4 != 0]
TEST3 = DF3[DF3['COMP_ID'] % 4 == 0]
```

```python
TRAIN3.shape
```

```python
TEST3.shape
```

# Step 23-3

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier(n_neighbors=7, metric='euclidean').fit(y=TRAIN3['CLASS'], X=TRAIN3.loc[:, VAR3])
```

# Step 23-4

```python
pred_prob = model.predict_proba(X=TEST3.loc[:, VAR3])
```

```python
pred_prob = pred_prob[:, 0]
```

```python
pred_prob.shape
```

```python
TEST3.head()
```

```python
TEST3.loc[:, 'PROBA'] = pred_prob
```

```python
TEST3.loc[:, 'PRED'] = model.predict(X=TEST3.loc[:, VAR3])
```

```python
TEST3.head()
```

```python
TEST3.groupby('PRED')['PROBA'].sum()
```

# Q 23 Answer

```python
round(603.857143, 3)
```
