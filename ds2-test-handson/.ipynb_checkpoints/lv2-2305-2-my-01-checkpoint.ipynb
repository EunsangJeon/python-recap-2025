### 1-1

```python
df_1_adm = df_1[df_1['ADMIT'] == 'Admitted']
```

```python
df_1_den = df_1[df_1['ADMIT'] == 'Denied']
df_1_den.shape
```

### 1-2

```python
df_1_adm['GRE'].rank(method='min', ascending=False)
```

```python
targets = ['GRE', 'TOEFL', 'CGPA']

for target in targets:
    df_1_adm[target + '_R'] = df_1_adm[target].rank(method='min', ascending=False)
    df_1_den[target + '_R'] = df_1_den[target].rank(method='min', ascending=False)
```

```python
df_1_den
```
