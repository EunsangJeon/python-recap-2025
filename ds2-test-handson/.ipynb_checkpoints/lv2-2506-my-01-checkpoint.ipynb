# 24-1

```python
from scipy.stats import pearsonr
```

```python
rates = []
targets = ['S1', 'S2', 'S3', 'S4', 'S5']

for target in targets:
    arr = df_24[target]
    rates.append(abs(pearsonr(arr, df_24['SCORE'])[0]))
```

```python
VAR24 = 'S2'
```

# 24-2

```python
A4 = df_24['IRATIO'].mean()
A4
```

```python
df_24.head()
```

```python
df_24['IHIGH'] = (df_24['IRATIO'] > A4) + 0
```

```python
df_24.head()
```

```python
df_24['IHIGH'].value_counts()
```

# 24-3

```python
df_high = df_24[df_24['IHIGH'] == 1]
df_high.shape
```

```python
df_low = df_24[df_24['IHIGH'] == 0]
df_low.shape
```

```python
B4 = abs(pearsonr(df_high[VAR24], df_high['SCORE'])[0])
B4
```

```python
C4 = abs(pearsonr(df_low[VAR24], df_low['SCORE'])[0])
C4
```

```python
D4 = abs(B4 - C4)
```

```python
print(round(D4, 3))
```

# Q25

```python
df_25 = CULTURE.copy()
```

```python
df_25.shape
```

```python
df_25.head()
```

# 25-1


```python
DF25 = df_25[df_25['FRATIO'] != -999]
```

```python
DF25.shape
```

# 25-2

```python
DF25.columns
```

```python
targets = ['NEMP', 'DNEMP','IRATIO', 'FRATIO', 'S1', 'S2', 'S3', 'S4', 'S5']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
for target in targets:
    value = StandardScaler().fit_transform(X=DF25[[target]])
    DF25[ target + '_S'] = value
```

```python
DF25.head()
```

# 25-3

```python
targets = ['NEMP_S', 'DNEMP_S', 'IRATIO_S', 'FRATIO_S', 'S1_S', 'S2_S', 'S3_S', 'S4_S', 'S5_S']
```

```python
targets
```

```python
for target in targets:
    DF25 = DF25[(DF25[target] <= 3) & (DF25[target] >= -3)]
```

```python
DF25.shape
```

# 25-4

```python
from sklearn.cluster import KMeans
```

```python
targets
```

```python
model = KMeans(n_clusters=3, random_state=1234).fit(X=DF25.loc[:, targets])
```

```python
DF25['LABEL'] = model.predict(X=DF25.loc[:, targets])
```

# 25-5

```python
DF25.groupby('LABEL')['SCORE'].mean()
```

```python
max(DF25.groupby('LABEL')['SCORE'].mean())
```

# Q25 Answer

```python
round(max(DF25.groupby('LABEL')['SCORE'].mean()), 2)
```

# Q 26

```python
df_26 = CULTURE.copy()
```

```python
df_26.shape
```

```python
df_26.head()
```

# 26-1

```python
TRAIN26 = df_26[df_26['C_ID'] % 5 != 0]
TEST26 = df_26[df_26['C_ID'] % 5 == 0]
```

```python
TRAIN26.shape
```

```python
TEST26.shape
```

# 26-2

```python
tmp = TRAIN26[TRAIN26['FRATIO'] != -999]
tmp.shape
```

```python
A6 = tmp['FRATIO'].mean()
A6
```

```python
TRAIN26['FRATIO'] = TRAIN26['FRATIO'].map(lambda x: A6 if (x == -999) else x)
```

```python
tmp = TRAIN26[TRAIN26['FRATIO'] != -999]
tmp.shape
```

```python
TEST26['FRATIO'] = TEST26['FRATIO'].map(lambda x: A6 if (x == -999) else x)
```

```python
tmp = TEST26[TEST26['FRATIO'] != -999]
tmp.shape
```

# 26-3

```python
TRAIN26.columns
```

```python
targets = ['NEMP', 'DNEMP', 'IRATIO', 'FRATIO', 'S1', 'S2', 'S3', 'S4','S5']
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
for target in targets:
    model = MinMaxScaler().fit(X=TRAIN26[[target]])
    TRAIN26[target + '_N'] = model.transform(X=TRAIN26[[target]])
    TEST26[target + '_N'] = model.transform(X=TEST26[[target]])
```

```python
TEST26.head()
```

```python
TRAIN26.head()
```

# 26-4

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
targets = ['NEMP_N', 'DNEMP_N', 'IRATIO_N', 'FRATIO_N', 'S1_N', 'S2_N', 'S3_N', 'S4_N', 'S5_N']
```

```python
targets
```

```python
model = KNeighborsRegressor(n_neighbors=9).fit(y=TRAIN26['SCORE'], X=TRAIN26.loc[:, targets])
```

```python
pred = model.predict(X=TEST26.loc[:, targets])
```

# 26-5

```python
from sklearn.metrics import mean_absolute_error
```

```python
mae = mean_absolute_error(y_true=TEST26['SCORE'], y_pred = pred)
mae
```

# Q26 Answer

```python
print(round(mae, 3))
```
