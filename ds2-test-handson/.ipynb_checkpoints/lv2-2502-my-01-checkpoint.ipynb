# Q25

```python
df_25 = CALL.copy()
```

```python
df_25.head()
```

# 25-1

```python
df_25['D_P_LEN'] = df_25['CALL_CNT'] * df_25['TM_AVG']
```

# 25-2

```python
df_25['AID'].value_counts()
```

```python
aids = []
for i in range(1,11):
    aids.append(i)
    
aids
```

```python
DF25 = pd.DataFrame()
```

```python
DF25['AID'] = aids
```

```python
DF25['A_CNT'] = df_25.groupby('AID')['CALL_CNT'].sum().reset_index()['CALL_CNT']
```

```python
DF25.head()
```

```python
DF25['A_LEN'] = df_25.groupby('AID')['D_P_LEN'].sum().reset_index()['D_P_LEN']
```

```python
DF25['A_PASS'] = df_25.groupby('AID')['PASS'].mean().reset_index()['PASS']
```

```python
DF25.head()
```

# 25-3

```python
from sklearn.preprocessing import StandardScaler
```

```python
targets = ['A_CNT', 'A_LEN']

for target in targets:
    DF25[target + '_S'] = StandardScaler().fit_transform(X=DF25.loc[:, [target]])
```

```python
DF25.head()
```

# 25-4

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(random_state=1234, n_clusters=2).fit(X=DF25.loc[:, ['A_CNT_S', "A_LEN_S"]])
```

```python
DF25['LABEL'] = model.predict(X=DF25.loc[:, ['A_CNT_S', "A_LEN_S"]])
```

```python
DF25.head()
```

# 25-5

```python
DF25.groupby('LABEL')['A_PASS'].mean()
```

```python
ANS = abs(0.459259 - 0.755556)
```

# Q25 Answer

```python
print(round(ANS, 3))
```
