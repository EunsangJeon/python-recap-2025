# Q2

### 2-1

```python
purchase.head()
```

```python
df2 = purchase.pivot_table(values=['PUR_AMT'], columns=['AGE_GRP'], index=['PRD_ID'], aggfunc='sum', fill_value=0).reset_index()
```

```python
df2.head()
```

```python
df2.columns = ['PRD_ID', 'AGE1_AMT', 'AGE2_AMT', 'AGE3_AMT', 'AGE4_AMT', 'AGE5_AMT']
```

```python
df2.head()
```

### 2-2

```python
from sklearn.preprocessing import StandardScaler
```

```python
for i in range(1,6):
    df2['AGE' + str(i) + '_AMT_Z'] = StandardScaler().fit_transform(X=df2[['AGE' + str(i) + '_AMT']])
```

```python
df2.head()
```

### 2-3

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=3, init='random', n_init=50,  random_state=1234).fit(X=df2.iloc[:, -5:])
```

```python
pred = model.predict(X=df2.iloc[:, -5:])
```

```python
pred
```

```python
df2['LABEL'] = pred
```

### 2-4

```python
df2.head()
```

```python
df2.groupby('LABEL')['AGE3_AMT'].mean()
```

### Q2 Answer

```python
round(45823.413084)
```
