```python
from sklearn.preprocessing import StandardScaler
```

### 2-2

### 2-3

```python
model_2 = KNeighborsClassifier(n_neighbors=2).fit(y=train_1['LABEL'], X=train_1.iloc[:, -24:])
model_3 = KNeighborsClassifier(n_neighbors=3).fit(y=train_1['LABEL'], X=train_1.iloc[:, -24:])
model_4 = KNeighborsClassifier(n_neighbors=4).fit(y=train_1['LABEL'], X=train_1.iloc[:, -24:])
```

```python
pred_2 = model_2.predict(X=test_1.iloc[:, -24:])
pred_3 = model_3.predict(X=test_1.iloc[:, -24:])
pred_4 = model_4.predict(X=test_1.iloc[:, -24:])
```

### 2-4

```python
from sklearn.metrics import accuracy_score
```

```python
as_2 = accuracy_score(y_true=test_1['LABEL'], y_pred = pred_2)
as_3 = accuracy_score(y_true=test_1['LABEL'], y_pred = pred_3)
as_4 = accuracy_score(y_true=test_1['LABEL'], y_pred = pred_4)
```

```python
as_2, as_3, as_4
```

# Q3

### 3-1

```python
df_3 = pd.read_csv('lv2-2305-4.csv')
df_3.head()
```

```python
df_3.shape
```

```python
iqr_arr = []
med_arr = []

for i in range(1,25):
    q1 = df_3['SENSOR' + str(i)].quantile(q = 0.25)
    q3 = df_3['SENSOR' + str(i)].quantile(q = 0.75)
    med = df_3['SENSOR' + str(i)].quantile(q = 0.5)
    iqr = q3 - q1
    iqr_arr.append(iqr)
    med_arr.append(med)
```

```python
for i in range(24):
    df_3 = df_3[(df_3['SENSOR' + str(i + 1)] > med_arr[i] - 2.3 * iqr_arr[i]) & (df_3['SENSOR' + str(i + 1)] < med_arr[i] + 2.3 * iqr_arr[i])]
```

```python
df_3.shape
```

### 3-2

```python
train = df_3[df_3['SEQ'] % 3 != 0]
test = df_3[df_3['SEQ'] % 3 == 0]
```

### 3-3

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier(max_depth=5, random_state=1205).fit(y=train['LABEL'], X=train.iloc[:, -24:])
```

### 3-4

```python
pred = model.predict(X=test.iloc[:, -24:])
```

```python
from sklearn.metrics import accuracy_score
```

```python
c = accuracy_score(y_true=test['LABEL'], y_pred=pred) * 100
c
```

### 3-5

```python
fi_arr = model.feature_importances_
```

```python
fi_df = pd.DataFrame(fi_arr, columns=['FI'])
```

```python
fi_df
```

```python
fi_df.sort_values(by='FI', ascending=False)
```

```python
target = ['SENSOR15', 'SENSOR4', 'SENSOR12', 'SENSOR7', 'SENSOR11', 'SENSOR13']
```

```python
model = DecisionTreeClassifier(max_depth=5, random_state=1205).fit(y=train['LABEL'], X=train.loc[:, target])
```

```python
pred = model.predict(X=test.loc[:, target])
```

### 3-6

```python
d = accuracy_score(y_true=test['LABEL'], y_pred=pred) * 100
d
```

```python
abs(c-d)
```

### Q3 Answer

```python
round(abs(c-d), 3)
```
