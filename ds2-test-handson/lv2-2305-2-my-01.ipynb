# Q2

```python
import pandas as pd
```

```python
df_2 = pd.read_csv('lv2-2305-2.csv')
```

```python
df_2.head()
```

```python
df_2.shape
```

### 2-1

```python
q1 = df_2['TOEFL'].quantile(0.25)
```

```python
q3 = df_2['TOEFL'].quantile(0.75)
```

```python
q1, q3
```

```python
iqr = q3 - q1
iqr
```

```python
df_2 = df_2[df_2['TOEFL'] >= (q1 - 1.5 * iqr)]
```

```python
df_2 = df_2[df_2['TOEFL'] <= (q3 + 1.5 * iqr)]
```

```python
df_2.shape
```

### 2-2

```python
df_2['ADMIT_S'] = (df_2['ADMIT'] == 'Admitted') + 0
```

```python
df_2.head()
```

### 2-3

```python
from sklearn.linear_model import LogisticRegression
```

```python
target_features = ['GRE', 'TOEFL', 'SOP', 'LOR', 'CGPA']
```

```python
model = LogisticRegression(solver='newton-cg', C=100000, random_state=1234).fit(y=df_2['ADMIT_S'], X= df_2.loc[:, target_features])
```

### 2-4

```python
pred = model.predict(df_2.loc[:, target_features])
```

```python
pred
```

```python
from sklearn.metrics import accuracy_score
```

### Q2 ANSWER

```python
round(accuracy_score(y_true=df_2['ADMIT_S'], y_pred=pred) * 100,2)
```

# Q3

### 3-1

```python
import pandas as pd
```

```python
df_q3 = pd.read_csv('lv2-2305-2.csv')
df_q3.head()
```

```python
df_q3 = df_q3[df_q3['ADMIT'] == 'Denied']
df_q3.head()
```

### 3-2

```python
dummies = pd.get_dummies(data=df_q3['UNIV_RATING'], drop_first=True, prefix='UR', dtype='int')
df_q3.loc[:, dummies.columns] = dummies.iloc[:,:]
```

```python
df_q3.head()
```

### 3-3

```python
target_features = ['GRE', 'TOEFL', 'SOP', 'LOR', 'CGPA']
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
df_q3.dtypes
```

```python
df_q3['GRE'] = df_q3['GRE'].astype('float')
df_q3['TOEFL'] = df_q3['TOEFL'].astype('float')
df_q3.dtypes
```

```python
for target in target_features:
    scaled = MinMaxScaler().fit_transform(X=df_q3.loc[:, [target]])
    df_q3.loc[:, [target]] = scaled
```

```python
df_q3.head()
```

### 3-4

```python
from sklearn.cluster import KMeans
```

```python
target_features = ['GRE', 'TOEFL', 'SOP', 'LOR', 'CGPA', 'RESEARCH', 'UR_B', 'UR_C', 'UR_D', 'UR_E']
```

```python
model_4 = KMeans(n_clusters=4, random_state=1234, n_init=50, max_iter=30).fit(X=df_q3.loc[:, target_features])
model_4.predict(X=df_q3.loc[:, target_features])
```

```python
model_5 = KMeans(n_clusters=5, random_state=1234, n_init=50, max_iter=30).fit(X=df_q3.loc[:, target_features])
model_5.predict(X=df_q3.loc[:, target_features])
```

```python
model_6 = KMeans(n_clusters=6, random_state=1234, n_init=50, max_iter=30).fit(X=df_q3.loc[:, target_features])
model_6.predict(X=df_q3.loc[:, target_features])
```

### 3-5

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(X=df_q3.loc[:, target_features], labels=model_4.predict(X=df_q3.loc[:, target_features]))
```

```python
silhouette_score(X=df_q3.loc[:, target_features], labels=model_5.predict(X=df_q3.loc[:, target_features]))
```

```python
silhouette_score(X=df_q3.loc[:, target_features], labels=model_6.predict(X=df_q3.loc[:, target_features]))
```
