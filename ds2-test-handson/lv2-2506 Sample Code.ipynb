```python
import pandas as pd
import numpy as np

CULTURE = pd.read_csv('lv2-2506.csv')
print(CULTURE.shape)
CULTURE
```

### Question 24

```python
# 24-1
DF24 = CULTURE.copy()
pvar = ['SCORE','S1','S2','S3','S4','S5']
cors = DF24[pvar].corr().loc['SCORE',['S1','S2','S3','S4','S5']]
print(cors)
VAR24 = abs(cors).idxmax()
print('VAR24:',VAR24)
```

```python
# 24-2
A4 = DF24.IRATIO.mean()
print('A4:',A4)
DF24['IHIGH'] = np.where(DF24['IRATIO']>A4, 1, 0)
```

```python
# 24-3
B4 = DF24[DF24['IHIGH']==1][['SCORE',VAR24]].corr().iloc[0,1]
C4 = DF24[DF24['IHIGH']==0][['SCORE',VAR24]].corr().iloc[0,1]
D4 = abs(B4-C4)
print('B4:',B4)
print('C4:',C4)
print('D4:',D4)
```

```python
# Answer 24
ans24f = D4
ans24 = round(ans24f, 3)

print(ans24f)
print(ans24)
```

### Question 25

```python
# 25-1
DF5 = CULTURE[CULTURE.FRATIO!=-999].copy().reset_index(drop=True)
DF5.shape
```

```python
# 25-2
from sklearn.preprocessing import StandardScaler
svar = ["NEMP", "DNEMP", "IRATIO", "FRATIO", "S1", "S2", "S3", "S4", "S5"]
svars = [v + '_S' for v in svar]
scaled = pd.DataFrame(StandardScaler().fit_transform(DF5[svar]), columns=svars)
DF25 = pd.concat([DF5, scaled], axis=1)
```

```python
# 25-3
DF25 = DF25[(abs(DF25[svars])>3).sum(axis=1)==0]
```

```python
# 25-4
from sklearn.cluster import KMeans
km = KMeans(random_state=1234, n_clusters=3).fit(DF25[svars])
DF25['CLST'] = km.labels_
```

```python
# 25-5
DF25.groupby('CLST').SCORE.mean()
```

```python
# Answer 25
ans25f = DF25.groupby('CLST').SCORE.mean().max()
ans25 = round(ans25f, 2)

print(ans25f)
print(ans25)
```

### Question 26

```python
# 26-1
TRAIN26 = CULTURE[CULTURE.C_ID%5!=0].copy().reset_index(drop=True)
TEST26 = CULTURE[CULTURE.C_ID%5==0].copy().reset_index(drop=True)
print(TRAIN26.shape, TEST26.shape)
```

```python
# 26-2
A6 = TRAIN26[TRAIN26.FRATIO!=-999].FRATIO.mean()
print('A6:',A6)

TRAIN26.loc[TRAIN26.FRATIO==-999, 'FRATIO'] = A6
TEST26.loc[TEST26.FRATIO==-999, 'FRATIO'] = A6
```

```python
# 26-3
from sklearn.preprocessing import MinMaxScaler
nvar = ["NEMP", "DNEMP", "IRATIO", "FRATIO", "S1", "S2", "S3", "S4", "S5"]
nvarn = [v + '_N' for v in nvar]
normer = MinMaxScaler().fit(TRAIN26[nvar])
TRAIN26n = pd.DataFrame(normer.transform(TRAIN26[nvar]), columns=nvarn)
TEST26n = pd.DataFrame(normer.transform(TEST26[nvar]), columns=nvarn)
```

```python
TRAIN26n
```

```python
# 26-4
from sklearn.neighbors import KNeighborsRegressor
mdl263 = KNeighborsRegressor(n_neighbors=9)
mdl263.fit(TRAIN26n, TRAIN26.SCORE)
pred263 = mdl263.predict(TEST26n)
```

```python
# 26-5
mae265 = np.mean(np.abs(pred263-TEST26.SCORE))
mae265
```

```python
# Answer 26
ans26f = mae265
ans26 = round(ans26f, 3)

print(ans26f)
print(ans26)
```
