# Q24

# 24-1

```python
df_24 = SNS.copy()
```

```python
df_24['UP_RATIO'] = df_24['UPVOTE'] / df_24['VIEW']
```

```python
df_24.head()
```

# 24-2

```python
df_24['CATEGORY'].value_counts()
```

```python
targets = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j',]
```

```python
from scipy.stats import pearsonr
```

```python
pearsonrs = []

for target in targets:
    df = df_24[df_24['CATEGORY'] == target]
    p = pearsonr(df['UP_RATIO'], df['REPUTATION'])[0]
    pearsonrs.append(p)
```

```python
pearsonrs
```

# 24-3

```python
min_p = min(pearsonrs)
max_p = max(pearsonrs)
gap_p = abs(min_p - max_p)

gap_p
```

# Q24 Answer

```python
print(round(gap_p, 3))
```

# Q25

```python
df_25 = SNS.copy()
```

# 25-1

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
targets = ['REPUTATION', 'VIEW']
```

```python
df_25.head()
```

```python
df_25.shape
```

```python
for target in targets:
    df_25[target + '_N'] = MinMaxScaler(feature_range=(0, 10)).fit_transform(X=df_25.loc[:, [target]])
```

```python
df_25.head()
```

# 25-2

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=4, random_state=1234).fit(X=df_25.loc[:, ['REPUTATION_N', 'VIEW_N']])
```

# 25-3

```python
centres = model.cluster_centers_
centres
```

# 25-4

```python
dists = []

for i in range(3):
    for j in range(i + 1, 4):
        dist = (((centres[i][0] - centres[j][0]) ** 2) + ((centres[i][1] - centres[j][1]) ** 2)) ** 0.5
        dists.append(dist)
```

```python
dists
```

# 25-5

```python
min_d = min(dists)
max_d = max(dists)
gap_d = abs(min_d - max_d)

gap_d
```

# Q25 Answer

```python
print(round(gap_d, 2))
```

# Q26

```python
df_26 = SNS.copy()
```

# 26-1

```python
targets = ['REPUTATION', 'REPLY', 'VIEW', 'UPVOTE']
```

```python
df_26.shape
```

```python
df_26.head()
```

```python
df_26['OUT'] = df_26['AID'].map(lambda x: 0)
df_26.head()
```

```python
for target in targets:
    q1 = df_26[target].quantile(0.25)
    q3 = df_26[target].quantile(0.75)
    iqr = q3 - q1
    df_26['OUT'] += (df_26[target] < q1 - 1.5 * iqr) + 0
    df_26['OUT'] += (df_26[target] > q3 + 1.5 * iqr) + 0
```

```python
df_26.head()
```

```python
DF6 = df_26[df_26['OUT'] == 0]
```

```python
DF6.shape
```

# 26-2

```python
DF6 = pd.get_dummies(DF6, columns=['CATEGORY'], drop_first=True)
```

```python
DF6.head()
```

# 26-3

```python
TRAIN6 = DF6[DF6['AID'] % 5 != 0]
TRAIN6.shape
```

```python
TEST6 = DF6[DF6['AID'] % 5 == 0]
TEST6.shape
```

# 26-4

```python
TRAIN6.shape
```

```python
TRAIN6.columns
```
