# Q3

### 3-1

```python
purchase.head()
```

```python
df3 = purchase.pivot_table(values=['PUR_AMT'], index=['USER_ID', 'GENDER', 'AGE_GRP', 'MARITAL_YN'], aggfunc='sum', fill_value=0).reset_index()
df3.head()
```

```python
df3_detail = purchase.pivot_table(values=['PUR_AMT'], columns=['PRD_CAT'], index=['USER_ID'], aggfunc='sum', fill_value=0).reset_index()
```

```python
df3_detail.head()
```

```python
df3_detail.columns = ['USER_ID', '1', '2', '3']
```

```python
df3_detail.head()
```

```python
df3_detail['MAX'] = df3_detail.loc[:, ['1', '2', '3']].idxmax(axis=1)
```

```python
df3_detail.head()
```

```python
df3['PRF_CAT'] = df3_detail['MAX']
```

```python
df3.head(10)
```

```python
df3.tail()
```

```python
df3['PRF_CAT'].value_counts()
```

### 3-2

```python
q2 = df3['PUR_AMT'].quantile(0.5)
q2
```

```python
q3 = df3['PUR_AMT'].quantile(0.75)
q3
```

```python
df3 = df3[(df3['PUR_AMT'] >= q2) & (df3['PUR_AMT'] <= q3)]
```

```python
df3.shape
```

```python
df3 = df3.reset_index(drop=True)
df3.head()
```

### 3-3

```python
df3.loc[:, ['GENDER_F', 'GENDER_M']] = pd.get_dummies(data=df3['GENDER'], dtype='int', prefix='GENDER')
df3.head()
```

```python
df3 = pd.concat([df3, pd.get_dummies(data=df3['AGE_GRP'], dtype='int', prefix='AGE_GRP')], axis=1)
```

```python
df3 = pd.concat([df3, pd.get_dummies(data=df3['MARITAL_YN'], dtype='int', prefix='MARITAL_YN')], axis=1)
```

```python
df3 = pd.concat([df3, pd.get_dummies(data=df3['PRF_CAT'], dtype='int', prefix='PRF_CAT')], axis=1)
```

```python
df3.shape
```

### 3-4

```python
df3 = df3.sort_values(by='USER_ID')
```

```python
df3.head()
```

```python
df3.tail()
```

```python
train3 = df3[df3['USER_ID'] % 4 != 0]
test3 = df3[df3['USER_ID'] % 4 == 0]
```

```python
train3.shape, test3.shape
```

### 3-5

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
model = DecisionTreeRegressor(random_state=1234).fit(X=train3.iloc[:, -11:], y=train3['PUR_AMT'])
```

### 3-6

```python
pred = model.predict(X=test3.iloc[:, -11:])
```

```python
from sklearn.metrics import mean_absolute_percentage_error
```

```python
mean_absolute_percentage_error(y_true=test3['PUR_AMT'], y_pred=pred)
```

### Q3 Answer

```python
round(mean_absolute_percentage_error(y_true=test3['PUR_AMT'], y_pred=pred) * 100, 2)
```
