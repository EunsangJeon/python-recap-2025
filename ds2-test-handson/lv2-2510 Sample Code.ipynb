```python
import pandas as pd
import numpy as np
```

```python
COMPONENT = pd.read_csv('lv2-2510.csv')
print(COMPONENT.info())
display(COMPONENT)
print(COMPONENT.CLASS.value_counts())
```

### Question 21.

```python
df1 = COMPONENT.copy()
print(df1.shape)
```

```python
# Step 21-1.

VAR1 = ['PERIM','MAJOR_LEN','MINOR_LEN','ASPECT','ECCENTRIC','SOLID','COMPACT']
VAR1_N = [c+'_N' for c in VAR1]
print(f'VAR1_N: {len(VAR1_N)},{VAR1_N}')
from sklearn.preprocessing import MinMaxScaler
n_df = pd.DataFrame(MinMaxScaler().fit_transform(df1[VAR1]), columns=VAR1_N)
df11 = pd.concat([df1,n_df], axis=1)
print(df11.shape)
```

```python
# Step 21-2.

from sklearn.cluster import KMeans
df11['CLS'] = KMeans(random_state=1234, n_clusters=3).fit_predict(df11[VAR1_N])
df11
```

```python
# Step 21-3.

tot_df = df11.groupby(['CLS'])['COMP_ID'].count().reset_index(name='CNT1')
display(tot_df)
seg_df = df11.groupby(['CLS','CLASS'])['COMP_ID'].count().reset_index(name='CNT2')
display(seg_df)
a_df = seg_df[seg_df.CLASS=='A'].reset_index(drop=True)
m_df = pd.merge(tot_df, a_df, on='CLS', how='left')
m_df['A_RATE'] = m_df['CNT2']/m_df['CNT1']
display(m_df)
```

```python
# Step 21-4.

A1 = np.max(m_df.A_RATE.tolist())
print(f'A1: {A1}')
B1 = df11[df11.CLS==m_df.loc[m_df.A_RATE==A1,'CLS'].values[0]]['AREA'].mean()
print(f'B1: {B1}')
```

```python
res1 = A1+B1
print(f'res1: {res1}')
ans1 = round(res1,3)
print(f'ans1: {ans1}')
```

### Question 22.

```python
df2 = COMPONENT.copy()
print(df2.shape)
```

```python
# Step 22-1.

X_VAR = ['AREA','PERIM','MAJOR_LEN','MINOR_LEN','ASPECT','ECCENTRIC','SOLID','COMPACT']
print(f'X_VAL: {len(X_VAR)}, {X_VAR}')
Y_VAR = 'CLASS'
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(random_state=1234, max_depth=3)
dtc.fit(df2[X_VAR],df2[Y_VAR])
```

```python
# Step 22-2.

f_df = pd.DataFrame(zip(X_VAR,dtc.feature_importances_),columns=['feat','imp'])
f_df = f_df.sort_values('imp', ascending=False).reset_index(drop=True)
A2 = sum(f_df.imp[:2])
print(f'A2: {A2}')
```

```python
f_df
```

```python
# Step 22-3.

pred = dtc.predict(df2[X_VAR])
actu = df2[Y_VAR]
B2 = sum((actu == 'A')&(pred == 'A')) / sum(actu == 'A')
print(f'B2: {B2}')
```

```python
res2 = A2+B2
print(f'res2: {res2}')
ans2 = round(res2,3)
print(f'ans2: {ans2}')
```

### Question 23.

```python
df3 = COMPONENT.copy()
print(df3.shape)
```

```python
# Step 23-1.

VAR3 = ['AREA','PERIM','MAJOR_LEN','MINOR_LEN','ASPECT','SOLID']
print(f'VAR3: {len(VAR3)},{VAR3}')
out_cnt = pd.Series(np.zeros(len(df3)))
for c in VAR3:
    Q1 = df3[c].quantile(0.25)
    Q3 = df3[c].quantile(0.75)
    IQR = Q3 - Q1
    LB = Q1 - 1.5*IQR
    UB = Q3 + 1.5*IQR
    out_tmp = np.where((df3[c]<LB) | (df3[c]>UB),1,0)
    print(f'c: {c}, out_no: {sum(out_tmp)}')
    out_cnt += out_tmp
DF3 = df3[out_cnt<2]
print(DF3.shape)
```

```python
# Step 23-2.

TRAIN3 = DF3[DF3.COMP_ID%4 != 0]
print(TRAIN3.shape)
TEST3 = DF3[DF3.COMP_ID%4 == 0]
print(TEST3.shape)
```

```python
# Step 23-3.

from sklearn.neighbors import KNeighborsClassifier
knc = KNeighborsClassifier(metric='euclidean', n_neighbors=7)
knc.fit(TRAIN3[VAR3], TRAIN3['CLASS'])
```

```python
# Step 23-4.

TEST3.loc[:,'pred'] = knc.predict(TEST3[VAR3])
TEST3.loc[:,'prob0'] = knc.predict_proba(TEST3[VAR3])[:,0]
TEST3.loc[:,'prob1'] = knc.predict_proba(TEST3[VAR3])[:,1]
actu = TEST3['CLASS']
prob_sum = sum(TEST3[TEST3.pred=='A']['prob0'])
print(f'prob_sum: {prob_sum}')
```

```python
res3 = prob_sum
print(f'res3: {res3}')
ans3 = round(res3,3)
print(f'ans3: {ans3}')
```
