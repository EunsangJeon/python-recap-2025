# Q1

```python
import pandas as pd
```

```python
df_1 = pd.read_csv('lv2-2305-3.csv')
df_1.head()
```

### 1-1

```python
shortened = []
for date in df_1['OBS_DT']:
    shortened.append(date[-8:])

shortened[:5]
```

```python
df_1['TIME'] = shortened
```

```python
df_1.head()
```

```python
df_1 = df_1[df_1['TIME'] >= df_1['TIME_SUNRISE']]
```

```python
df_1 = df_1[df_1['TIME'] <= df_1['TIME_SUNSET']]
```

```python
df_1.head()
```

### 1-2

```python
features = ['PRESSURE', 'HUMIDITY', 'WIND_SPEED']
target = 'RADIATION'
```

```python
from scipy.stats import pearsonr
```

```python
pearsons = []

for feature in features:
    pearsons.append(abs(pearsonr(x=df_1[feature], y=df_1[target])[0]))
```

```python
pearsons
```

### Q1 Answer

```python
round(max(pearsons), 3)
```

# Q3

```python
df_3 = pd.read_csv('lv2-2305-3.csv')
df_3.head()
```

### 3-1

```python
dates = []

for datetime in df_3['OBS_DT']:
    dates.append(datetime[:10])

dates[:10]

df_3['DATE'] = dates

df_3.head()
```

```python
sunrises = df_3['TIME_SUNRISE']
sunsets = df_3['TIME_SUNSET']
```

```python
len(sunrises), len(sunsets)
```

```python
sun_dur = []

for i in range(32686):
    sun_dur.append(3600 * (int(sunsets[i][:2]) - int(sunrises[i][:2])) + 60 * (int(sunsets[i][3:5]) - int(sunrises[i][3:5])) + (int(sunsets[i][6:8]) - int(sunrises[i][6:8])))
```

```python
len(sun_dur)
```

```python
sun_dur[:10]
```

```python
df_3['SUN_DURATION'] = sun_dur
df_3.head()
```

```python
df_3.loc[df_3['DATE'] == '2022-09-01']
```

```python
df_3.groupby('DATE')[['SUN_DURATION']].mean()
```

### 3-2

```python
df_sum = df_3.groupby('DATE')[['SUN_DURATION', 'TEMPERATURE', 'PRESSURE', 'HUMIDITY', 'WIND_DIRECTION', 'WIND_SPEED', 'RADIATION']].mean().reset_index()
```

```python
df_sum.head()
```

```python
df_sum.loc[df_sum['DATE'] == '2022-09-29']
```

### 3-3

```python
df_sum.shape
```

```python
df_sum.head()
```

```python
len(df_sum['DATE'])
```

```python
df_sum['day'] = df_sum['DATE'].map(lambda x: x[-2:])
```

```python
df_sum.head()
```

```python
df_train = df_sum.loc[(df_sum['day'] != '01') & (df_sum['day'] != '15')]
```

```python
df_train.shape
```

```python
df_test = df_sum.loc[(df_sum['day'] == '01') | (df_sum['day'] == '15')]
```

```python
df_test.shape
```

### 3-4

```python
from sklearn.linear_model import LinearRegression
```

```python
df_train.iloc[:, 7]
```

```python
model = LinearRegression().fit(y=df_train.iloc[:, 7], X=df_train.iloc[:, 1:7])
```

### 3-5

```python
pred = model.predict(X=df_test.iloc[:, 1:7])
```

```python
from sklearn.metrics import mean_squared_error
```

### Q3 Answer

```python
round(mean_squared_error(y_true=df_test.iloc[:, 7], y_pred=pred), 2)
```
