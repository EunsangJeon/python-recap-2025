# Q22

```python
df_22 = JOB.copy()
```

```python
df_22.shape
```

```python
df_22.head()
```

```python
df_22['EXP_LEVEL'].value_counts()
```

# Step 22-1

```python
exps = []

for exp in df_22['EXP_LEVEL']:
    if (exp == 'Entrylevel') | (exp == 'Midlevel'):
        exps.append('L')
    else:
        exps.append('H')
```

```python
df_22['EXP'] = exps
```

```python
df_22.head()
```

# Step 22-2

```python
cats = ['A', 'B', 'C', 'D', 'E', 'F', 'G']
```

```python
res = []

for cat in cats:
    df = df_22[df_22['JOB_CATEGORY'] == cat]
    df_h = df[df['EXP'] == 'H']
    df_l = df[df['EXP'] == 'L']

    res.append((abs(df_h['SALARY'].mean() - df_l['SALARY'].mean())) / (((len(df_h) * df_h['SALARY'].std()) + (len(df_l) * df_l['SALARY'].std())) / (len(df_h) + len(df_l))))
```

```python
len(res)
```

```python
res
```

# Step 22-3

```python
A2 = max(res)
B2 = min(res)
```

# Q22 Answer

```python
round(abs(A2 - B2), 3)
```

# Q23

```python
df_23 = JOB.copy()
```

```python
df_23.shape
```

```python
df_23.head()
```

# Step 23-1

```python
targets = ['JOB_CATEGORY', 'EXP_LEVEL', 'WORK_SETTING', 'COMP_SIZE']
```

```python
df_23 = pd.get_dummies(data=df_23, drop_first=True, columns=targets, dtype='int')
```

```python
df_23.head()
```

# Step 23-2

```python
TRAIN30 = df_23[df_23['E_ID'] % 5 != 0]
TEST30 = df_23[df_23['E_ID'] % 5 == 0]
```

```python
TRAIN30.shape
```

```python
TEST30.shape
```

# Step 23-3

```python
q1 = TRAIN30['SALARY'].quantile(q=0.25)
q3 = TRAIN30['SALARY'].quantile(q=0.75)
iqr = q3 - q1

TRAIN31 = TRAIN30[(TRAIN30['SALARY'] >= q1 - 1.5 * iqr) & (TRAIN30['SALARY'] <= q3 + 1.5 * iqr)]
TRAIN31.shape
```

# Step 23-4

```python
TRAIN31.head()
```

```python
TRAIN31.shape
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
model = DecisionTreeRegressor(max_depth=4, min_samples_split=100, random_state=1234).fit(X=TRAIN31.iloc[:, -11:], y=TRAIN31['SALARY'])
```

# Step 23-5

```python
pred = model.predict(X=TEST30.iloc[:, -11:])
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mse = mean_squared_error(y_true=TEST30['SALARY'], y_pred=pred)
```

```python
mse
```

# Q23 Answer

```python
print(round(mse ** 0.5 , 2))
```
