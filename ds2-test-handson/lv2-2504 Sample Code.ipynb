```python
import pandas as pd
import numpy as np
```

```python
SNS = pd.read_csv('lv2-2504.csv')
SNS.info()
```

# 24

```python
df240 = SNS.copy()
```

24-1

```python
df240['UP_RATIO'] = df240['UPVOTE']/df240['VIEW']
df240['UP_RATIO'].describe()
```

24-2

```python
list_cor = []
for cat in df240['CATEGORY'].unique():
    cor00 = df240.loc[df240['CATEGORY']==cat, ['UP_RATIO','REPUTATION']].corr().iloc[0, 1]
    list_cor.append(cor00)
list_cor
```

24-3

```python
res24 = np.max(list_cor) - np.min(list_cor)
ans24 = round(res24, 3)
print(res24)
print('ans24 : ', ans24)
```

# 25

```python
df250 = SNS.copy()
```

25-1

```python
for col in ['REPUTATION','VIEW']:
    max00 = df250[col].max()
    min00 = df250[col].min()
    str00 = col + '_N'
    df250[str00] = 10*(df250[col]-min00)/(max00-min00)
```

25-2

```python
from sklearn.cluster import KMeans
km00 = KMeans(random_state=1234, n_clusters=4)
km00.fit(df250.loc[:, ['REPUTATION_N','VIEW_N']])
srs00 = pd.Series(km00.labels_)
srs00.index = df250.index
df250['label'] = srs00
```

25-3

```python
list_cent = []
for lab in range(4):
    cent00 = np.array(df250.loc[df250['label']==lab,['REPUTATION_N','VIEW_N']].mean(axis=0))
    list_cent.append(cent00)
```

```python
list_cent
```

25-4

```python
list_dist = []
for i in range(4):
    for j in range(i+1, 4):
        cent00 = list_cent[i]
        cent01 = list_cent[j]
        dist00 = np.sqrt(np.sum((cent00 - cent01)**2))
        list_dist.append(dist00)
list_dist
```

25-5

```python
res25 = np.max(list_dist) - np.min(list_dist)
ans25 = round(res25, 2)
print(res25)
print('ans25 : ', ans25)
```

# 26

```python
df260 = SNS.copy()
```

26-1

```python
df261 = df260.copy()
for col in ['REPUTATION','REPLY','VIEW','UPVOTE']:
    q1 = df260[col].quantile(0.25)
    q3 = df260[col].quantile(0.75)
    iqr = q3 - q1
    df261 = df261[(df261[col]>=q1 - 1.5*iqr) & (df261[col]<=q3 + 1.5*iqr)]
df261.shape
```

26-2

```python
df_dum = pd.get_dummies(df261['CATEGORY'], drop_first=True)
list_dum = list(df_dum.columns)
DF6 = pd.concat([df261, df_dum], axis=1)
```

26-3

```python
train6 = DF6[DF6['AID']%5!=0].copy()
test6 = DF6[DF6['AID']%5==0].copy()
print(train6.shape, test6.shape)
```

26-4, 26-5

```python
list_fts = ['REPLY','VIEW','UPVOTE'] + list_dum
from sklearn.linear_model import LinearRegression
LR00 = LinearRegression()
LR00.fit(train6.loc[:, list_fts], train6['REPUTATION'])
A6 = np.mean(LR00.predict(test6.loc[:, list_fts]))
res26 = A6
ans26 = round(res26, 3)
print(res26)
print('ans26 : ',ans26)
```
