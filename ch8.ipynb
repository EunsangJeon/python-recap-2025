# Exception Handling

```python
def sample_function(n):
    try:
        print(15 / n)
    except:
        pass
```

```python
sample_function(0)
```

```python
def sample_function_2(n):
    try:
        print(15 / n)
    except ZeroDivisionError:
        pass
```

```python
sample_function_2(0)
```

```python
def sample_function_3(n):
    try:
        print(15 / n)
    except ZeroDivisionError:
        raise RuntimeError("ERROR")
```

```python
sample_function_3(0)
```

```python
def sample_function_4(n):
    try:
        print(15 / n)
    except ZeroDivisionError:
        pass
    finally:
        print('hew')
```

```python
sample_function_4(0)
```
